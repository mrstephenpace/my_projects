This script loads the COCO dataset (2017) instances validation json script. It filters out only the images with people ('person') and downloads the first 500 of these images. It also converts the instances json file for the 500 images into a .csv file. This is used when converting the images to a TensorFlow record.

Lastly, this script loads the COCO dataset (2017) instances training json script. It filters out only the images with people ('person') and downloads the first 5000 of these images. It also converts the instances json file for the 5000 images into a .csv file. This is used when converting the images to a TensorFlow record.


This script can be modified to increase/decrease validation/training dataset sizes as neccessary.

```python
from pycocotools.coco import COCO
annFile = 'annotations/instances_val2017.json'
```

```python
coco = COCO(annFile)
cats = coco.loadCats(coco.getCatIds())

catIds = coco.getCatIds(catNms=['person'])
imgIds = coco.getImgIds(catIds=catIds)
imgIds = imgIds[0:499]
images = coco.loadImgs(imgIds)
```

```python
len(images)
```

```python
import requests
import os
for im in images:
    print("im: ", im['file_name'])
    img_data = requests.get(im['coco_url']).content
    with open(os.path.join('images/test',im['file_name']), 'wb') as handler:
        handler.write(img_data)
```

```python
import csv
classes = 'person'
with open('data/person_test.csv', mode='w', newline='') as annot:
    annot_writer = csv.writer(annot)
    annot_writer.writerow(['filename',
                           'width',
                           'height',
                           'class',
                           'xmin',
                           'ymin',
                           'xmax',
                           'ymax'])
    for im in images:
        annIds = coco.getAnnIds(imgIds=im['id'], catIds=catIds, iscrowd=None)
        anns = coco.loadAnns(annIds)
        for i in range(len(anns)):
            annot_writer.writerow([im['file_name'],im['width'],im['height'],classes,int(round(anns[i]['bbox'][0])),int(round(anns[i]['bbox'][1])),int(round(anns[i]['bbox'][0] + anns[i]['bbox'][2])),int(round(anns[i]['bbox'][1] + anns[i]['bbox'][3]))])
annot.close()
```

```python
annFile = 'annotations/instances_train2017.json'
```

```python
coco = COCO(annFile)
cats = coco.loadCats(coco.getCatIds())

catIds = coco.getCatIds(catNms=['person'])
imgIds = coco.getImgIds(catIds=catIds)
imgIds = imgIds[0:4999]
images = coco.loadImgs(imgIds)
```

```python
for im in images:
    print("im: ", im['file_name'])
    img_data = requests.get(im['coco_url']).content
    with open(os.path.join('images/train',im['file_name']), 'wb') as handler:
        handler.write(img_data)
```

```python
classes = 'person'
with open('data/person_train.csv', mode='w', newline='') as annot:
    annot_writer = csv.writer(annot)
    annot_writer.writerow(['filename',
                           'width',
                           'height',
                           'class',
                           'xmin',
                           'ymin',
                           'xmax',
                           'ymax'])
    for im in images:
        annIds = coco.getAnnIds(imgIds=im['id'], catIds=catIds, iscrowd=None)
        anns = coco.loadAnns(annIds)
        for i in range(len(anns)):
            annot_writer.writerow([im['file_name'],
                                   im['width'],
                                   im['height'],
                                   classes,
                                   int(round(anns[i]['bbox'][0])),
                                   int(round(anns[i]['bbox'][1])),
                                   int(round(anns[i]['bbox'][0] + anns[i]['bbox'][2])),
                                   int(round(anns[i]['bbox'][1] + anns[i]['bbox'][3]))])
annot.close()
```
